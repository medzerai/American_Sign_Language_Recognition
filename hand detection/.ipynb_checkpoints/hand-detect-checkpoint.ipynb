```python
import cv2
from cvzone.HandTrackingModule import HandDetector
from PIL import Image as im
import keyboard
import numpy as np
```

```python
cap = cv2.VideoCapture(0)
key = cv2.waitKey(1)
detector = HandDetector(detectionCon=0.8,maxHands=2)
```

```python
while True:
    
    success,img=cap.read()
    dim=img.shape
    imm=img.copy()
    hands,img=detector.findHands(img)
    #hands,img=detector.findHands(img,draw=False) #no draw
    
    #hand -dict {lmList -bbox -center -type}
    if hands:
        hand1=hands[0]
        lmList1=hand1["lmList"] # list of 21 Landmarks points for each finger..
        bbox1=hand1["bbox"] # bounding box x,y,w,h
        centerPoint1= hand1["center"] #center of the hand x,y
        handType1=hand1["type"]
        
        fingers1 = detector.fingersUp(hand1)
        
        if(len(hands)==2):
            hand2=hands[1]
            lmList2=hand2["lmList"] # list of 21 Landmarks points for each finger..
            bbox2=hand2["bbox"] # bounding box x,y,w,h
            centerPoint2= hand2["center"] #center of the hand x,y
            handType2=hand2["type"]
            
            fingers2 = detector.fingersUp(hand2)
            
#             print(fingers2)
#             print(handType2)
            
#             length,info,img = detector.findDistance(lmList1[8],lmList2[8],img)
#             if handType2=="Left":
#                 if fingers2 == [0,1,0,0,0] : 
#                     b1=[]
#                     for k in bbox1:
#                         if k<0:
#                             b1.append(0)
#                         b1.append(k)
#                     o=50
#                     imk = imm[b1[1]-o:b1[1]+b1[3]+o, b1[0]-o:b1[0]+b1[2]+o]
#                     data = im.fromarray(imk)
#                     data.save('./dataHands/opencv'+str(np.random.randint(100))+'.png')
#             elif handType1=="Left":
#                 if fingers1 == [0,1,0,0,0] : 
#                     b1=[]
#                     for k in bbox2:
#                         if k<0:
#                             b1.append(0)
#                         b1.append(k)
#                     o=50
#                     imk = imm[b1[1]-o:b1[1]+b1[3]+o, b1[0]-o:b1[0]+b1[2]+o]
#                     data = im.fromarray(imk)
#                     data.save('./dataHands/opencv'+str(np.random.randint(100))+'.png')


        # x,y,w,h
        xb,yb,wb,hb=bbox1
        if keyboard.is_pressed('s'): 
            o=20
            xb,yb,wb,hb=bbox1
            if wb<hb:
                wb=int(wb+((hb-wb)/2))
                xb=int(xb-((hb-wb)/2))
            elif hb>wb:
                hb=int(hb+((wb-hb)/2))
                yb=int(yb-((wb-hb)/2))
                    
            if xb-o<0:
                xb=0
            else:
                if xb+wb+o>dim[1]:
                    wb=dim[1]-xb
                
            if yb-o<0:
                yb=0
            else:
                if yb+hb+o>dim[0]:
                    hb=dim[0]-yb
                
            b1=[xb,yb,wb,hb]
            
            
            imk = imm[b1[1]-o:b1[1]+b1[3]+o, b1[0]-o:b1[0]+b1[2]+o]
            data = im.fromarray(imk)
            data.save('./dataHands/opencv'+str(np.random.randint(100))+'.png')

            
    cv2.imshow("Image",img)
    cv2.waitKey(1)
```
